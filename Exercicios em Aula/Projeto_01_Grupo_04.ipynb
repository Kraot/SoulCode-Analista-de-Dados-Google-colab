```python
#Aqui é a nossa biblioteca ja com alguns arquivos salvos dentro dela.
funcionarios = {1: {'nome': 'Matheus', 'cargo': 'Consultor', 'cidade': 'Rj', 'estado': 'Rj', 'idade': 20, 'escolaridade': 'Superior', 'salario': 1000.0}, 2: {'nome': 'Kellen', 'cargo': 'Analista', 'cidade': 'Sp', 'estado': 'Sp', 'idade': 20, 'escolaridade': 'Superior', 'salario': 1000.0}} # Criando um dicionário vazio


#Implementamos uma Def para  Adicionar Funcionarios a nossa Biblioteca
def adicionar_funcionarios():
  while True: #fizemos um tratamento de exeção para que usuario não possa escrever um id ja existente e também não possa escrever um id que não seja um int.
    try:   # aqui adicionar int para ser inserido um numero inteiro visando corrigir possiveis erros de digitaçao
      id = int(input('Informe o ID do funcionário: '))  # cada input serve para usuario colocar as informações pedidas.
      if id in funcionarios:
        print()
        print('Esse id já existe tente outro')
        print()
      else:
        break #Fim de função
    except:
      print()
      print('O valor informado não é um número')
      print()
  print('-------------Insira os dados do funcionario solicitado abaixo:---------')
  nome = input('Nome : ')
  cargo = input('Cargo : ')
  cidade = input('Cidade : ')
  estado = input('Estado : ')
  while True:  #fizemos um tratamento de exceção para que usuario não possa escrever uma variavel que não seja um numero inteiro
    try:# aqui adicionar int para ser inserido um numero inteiro visando corrigir possiveis erros de digitaçao.
       idade = int(input('Idade: '))
       break
    except:
      print()
      print('O valor informado não é um número')
      print()

  escolaridade = input('Escolaridade : ')

  while True: #fizemos um tratamento de exceção para que usuario não possa escrever uma variavel que não seja um int
    try:# aqui adicionar float para ser inserido um numero inteiro visando corrigir possiveis erros de digitaçao
       salario = float(input('Salário : '))
       break
    except:
      print()
      print('O valor informado não é um número')
      print()

  # aqui insirimos os imputs dentro do dicionario
  funcionarios[id] = {'nome': nome, 'cargo': cargo, 'cidade': cidade, 'estado': estado, 'idade': idade, 'escolaridade': escolaridade, 'salario': salario}
  print()
  print('-----------------------------Funcionario cadastrado com sucesso----------------------------')
  print()
#--------------------------------------------------------------------------------

def atualizar_funcionarios():
  while True:#fizemos um tratamento de exceção para que usuario não possa escrever uma variavel que não seja um int
    try: # aqui adicionar int para ser inserido um numero inteiro visando corrigir possiveis erros de digitaçao.
       id = int(input('Informe o ID do funcionário que deseja atualizar: '))
       break
    except:
      print()
      print('O valor informado não é um número')
      print()


  if id in funcionarios:
    print()
    print('-----------------------Funcionário encontrado--------------------------')
    print()
    print('-------------Insira os dados do funcionario solicitado abaixo:---------')
    nome = input('Nome : ')
    cargo = input('Cargo : ')
    cidade = input('Cidade : ')
    estado = input('Estado : ')
    while True:#fizemos um tratamento de exceção para que usuario não possa escrever uma variavel que não seja um int
     try: #aqui adicionar int para ser inserido um numero inteiro visando corrigir possiveis erros de digitaçao.
       idade = int(input('Idade : '))
       break
     except:
      print()
      print('O valor informado não é um número')
      print()
    escolaridade = input('Escolarida : ')
    while True:
     try:# aqui adicionar float para ser inserido um numero inteiro visando corrigir possiveis erros de digitaçao
       salario = float(input('Salário :'))
       break
     except:
      print()
      print('O valor informado não é um número')
      print()

    funcionarios[id] = {'nome': nome, 'cargo': cargo, 'cidade': cidade, 'estado': estado, 'idade': idade, 'escolaridade': escolaridade, 'salario': salario}
    print()
    print('----------Funcionario atualizado com sucesso-------------')
    print()

  else:
    print()
    print('--------Usuário não encontrado-----------')
    print()

#---------------------------------------------------------------------------------

def exibir_funcionario():
  while True:
    try:
       id = int(input('Informe o ID do funcionário que deseja exibir: '))
       break
    except:
      print()
      print('o valor informado não é um número')
      print()

  if id in funcionarios:#condicional para exibir o funcionário que foi solicitado
      d = funcionarios[id]#acessa dados do funcionário com base no id
      print()
      print("ID:", id)#imprime a id
      print("Nome:", d['nome'])#imprime os dados armazenados na variável d
      print("Cargo:", d['cargo'])
      print("Cidade:", d['cidade'])
      print("Estado:", d['estado'])
      print("idade:", d['idade'])
      print("Escolaridade:", d['escolaridade'])
      print("Salário:", d['salario'])
      print('_____________________________________')
  else: #caso a condição acima não seja realizada, irá imprimir Funcionário não encontrado
    print()
    print('-------Funcionário não encontrado--------')
    print()
#---------------------------------------------------------------------------------

def listar_funcionarios():
  if funcionarios:
    for id, d in funcionarios.items():#itera os valores dos itens no dicionário funcionarios e faz estrutura de repetição
      print("ID:", id)#imprime a id
      print("Nome:", d['nome']) #imprime os dados armazenados na variável d
      print("Cargo:", d['cargo'])
      print("Cidade:", d['cidade'])
      print("Estado:", d['estado'])
      print("idade:", d['idade'])
      print("Escolaridade:", d['escolaridade'])
      print("Salário:", d['salario'])
      print('_________________________________')
  else:#caso a condição acima não seja realizada, irá imprimir Nenhum funcionário cadastrado
    print()
    print('------Nenhum funcionário cadastrado-------')
    print()
#---------------------------------------------------------------------------------
def del_funcionarios():
   id = int(input('Informe o ID do funcionário que deseja DELETAR: '))
   if id in funcionarios:#condicional que se id estiver dentro de funcionários, faz o comando de deletar o id
        del funcionarios[id]
        print()
        print('----Funcionário DELETADO com sucesso!!!!------')
        print()
   else:#caso a condição acima não seja realizada, irá imprimir Funcionário não encontrado
      print()
      print('-------Funcionário não encontrado--------')
      print()
#--------------------------------------------------------------------------------
while True: #estrutura de repetição para repetir o programa até que o usuário insira a opção 6 (sair)
   print('================= Sistemas_De_Cadastro ====================')
   print('Seleciona a ação que deseja executar:')
   print('(1) Adicionar funcionário ')
   print('(2) Atualizar funcionário ')
   print('(3) Exibir funcionário ')
   print('(4) Listar funcionarios ')
   print('(5) Excluir funcionário ')
   print('(6) Se você deseja sair ')
   print('-----------------------')
   D = input('Escolha uma opção: ') #entrada para o usuário digitar uma das opções acima
   print('-----------------------')
   print('================= Sistemas_De_Cadastro ====================')


   if D == '1':#se a condicional for verdadeira (inserir 1), irá iniciar a def adicionar_funcionarios
      adicionar_funcionarios()
   elif D == '2':#se a condicional for verdadeira, irá iniciar a def atualizar_funcionarios
      atualizar_funcionarios()
   elif D == '3':#se a condicional for verdadeira, irá iniciar a def exibir_funcionarios
      exibir_funcionario()
   elif D == '4':#se a condicional for verdadeira, irá iniciar a def listar_funcionarios
      listar_funcionarios()
   elif D == '5':#se a condicional for verdadeira, irá iniciar a def del_funcionarios
      del_funcionarios()
   elif D == '6':#se a condicional for verdadeira, irá sair do programa
     print('Saindo do programa....')
     break
   else:#irá dizer que o valor é inválido caso o usuário não digite nenum dos valores especificados acima
    print(f'Valor {D} invalido Digite (1, 2, 3, 4, 5 ou 6)')
```
